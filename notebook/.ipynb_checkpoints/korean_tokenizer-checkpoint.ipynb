```python
import json

import pandas as pd

with open('NEtaggedCorpus_train.json', 'r') as f:
    corpus = f.read()
data = json.loads(corpus)['sentence']
```

```python
objs = []
for sentence in data:
    obj = dict()
    morp = sentence['morp']
    obj['tokens'] = [m['lemma'] for m in morp]
    obj['labels'] = [m['type'] for m in morp]
    assert(len(obj['tokens']) == len(obj['labels']))
#     print(obj)
    objs.append(obj)
print(objs)
```

```python
with open("train.json", "w", encoding='UTF-8') as f:
    json.dump(objs, f, ensure_ascii=False)
```
